```python
def homework1(url_winequality_data):
#問題のデータを表にする
    df = pd.read_csv(url_winequality_data, sep=';') #sepは区切り文字
    df2 = df.sort_values(by=["total sulfur dioxide"], ascending=True) #total sulfur dioxideについて昇順(小さい順)で並び替え
 #初期値を入力aは全体のデータ数、mはm分位数とする
    a = df2.shape[0] #全体のデータ数(0列目にはデータ数、1列目にはカラム(データの種類)数が入っている)
    m = 5 #分位点で5つのグループに分ける（つまり、五分位数を考える）
    n = int(((a+1)/m)-1) #各グループのデータ数 #数値解的に算出 
    ph = np.array([]) #空のndarrayを作る
    for i in range(1, m+1): #各グループのphの平均値を求める
      ph_v=df2['pH'][(n+1)*(i-1):i*n+(i-1)].mean() #各グループの先頭から最後 #g1[0:319],g2[320:639],g3[640:959],g4[960:1279],g5[1280:1599]
      ph=np.append(ph, ph_v)
    my_result = ph.max()+ph.min() #各グループのphの平均値の一番小さいものと一番大きいものを足す
    return ph, my_result
```

```python
def homework2(url_winequality_data):
#問題のデータを表にする
    df = pd.read_csv(url_winequality_data, sep=';') #sepは区切り文字
    df2 = df.sort_values(by=["total sulfur dioxide"], ascending=True) #total sulfur dioxideについて昇順(小さい順)で並び替え
 #初期値を入力aは全体のデータ数、mはm分位数とする
    a = df2.shape[0] #全体のデータ数(0列目にはデータ数、1列目にはカラム(データの種類)数が入っている)#1599
    m = 5 #分位点で5つのグループに分ける（つまり、五分位数を考える）
    n = int(((a+1)/m)-1) #各グループのデータ数 #数値解的に算出#319*5グループ+4つの分位数=1599 
    ph = np.array([]) #空のndarrayを作る
    for i in range(1, m+1): #各グループのphの平均値を求める
      if i==1:
        ph_v=df2['pH'][0:i*n+(i-1)].mean()
      else:
        ph_v=df2['pH'][(n+1)*(i-1)+1:i*n+(i-1)].mean() #各グループの先頭から最後 #g1[0:319],第1五分位数：320,g2[321:639],第2五分位数：640,g3[641:959],第3五分位数：960,g4[961:1279],第4五分位数：1280,g5[1281:1599]
      ph=np.append(ph, ph_v)
    my_result = ph.max()+ph.min() #各グループのphの平均値の一番小さいものと一番大きいものを足す
    return ph, my_result
```

```python
def homework3(url_winequality_data):
    #CSVをDFに読み込む
    df = pd.read_csv(url_winequality_data ,sep=";")

    #5つに分割して、ラベルを追加する。
    df['tsq'] = pd.qcut(df['total sulfur dioxide'], 5, labels=["dv1","dv2","dv3","dv4","dv5"])

    #グループごとに平均をとる
    df1 = df.groupby('tsq').mean()  

    #最大と最小を足す
    my_result = df1['pH'].max()+df1['pH'].min()

    #値を返す
    return my_result
```
