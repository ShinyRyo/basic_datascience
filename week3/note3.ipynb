```
# working place. everything 
def homework(target_online_retail_data_tb):
    #CustomerID毎の合計金額を出す
    df = target_online_retail_data_tb #元データ
    dfCus = df.groupby('CustomerID') #CustomerIDでまとめたオブジェクトを作る
    #dfCus.sum()['TotalPrice] #この形だとseriesになる #dfCus.sum()だとQuantityとUnitPriceも付いてくる
     #新しいdfの要素（data, index, column）
    list1=dfCus.sum()['TotalPrice'].values 
    index1 =dfCus.sum().index
    columns1 =["TotalPriceC"]
    dfN=pd.DataFrame(data=list1, index=index1, columns=columns1) #Customer毎にまとめたdfを作成
    #グルーピング(新しいdf(dfN)に対して)
    deciles=["decile01","decile02","decile03","decile04","decile05","decile06","decile07","decile08","decile09","decile10"]
    bins=10
    dfN['decile'] = pd.qcut(dfN['TotalPriceC'], bins, labels=deciles) #dfNに新しいカラムを加える
    #グループごとに購入金額の合計をとる
    dfP = dfN.groupby('decile').sum() #groupオブジェクトに.sum()などを加えるとデータフレームになる
    #グループ毎の割合を出す
    dfP['share']=dfP['TotalPriceC']/df['TotalPrice'].sum()
    my_result = dfP['share']
    return my_result
```
